```python
# Q1: Data Analysis and Visualization

# Part I: Randomized Sales Data Generation (NumPy)

# Import necessary libraries
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Step 1: Initialize Random Seed
# Using roll number 102317201 as the random seed for unique data generation
random_seed = 102317201
np.random.seed(random_seed)

# Step 2: Generate Sales Data
# Generate a NumPy array of random sales data for 12 months and 4 categories
sales_data = np.random.randint(1000, 5000, size=(12, 4))

# Define category names and month labels
categories = ['Electronics', 'Clothing', 'Home & Kitchen', 'Sports']
months = ['Jan', 'Feb', 'Mar', 'Apr', 'May', 'Jun', 'Jul', 'Aug', 'Sep', 'Oct', 'Nov', 'Dec']

# Step 3: Convert to DataFrame
# Convert the NumPy array into a Pandas DataFrame with appropriate labels
sales_df = pd.DataFrame(sales_data, columns=categories, index=months)

# Display the first few rows of the DataFrame
print("Sales Data (First 5 Rows):")
print(sales_df.head())

# Part II: Data Manipulation and Analysis (Pandas)

# Step 1: Display Summary Statistics
print("\nSummary Statistics:")
print(sales_df.describe())

# Step 2: Calculate Total Sales per Category and per Month
total_sales_per_category = sales_df.sum(axis=0)
total_sales_per_month = sales_df.sum(axis=1)

print("\nTotal Sales Per Category:")
print(total_sales_per_category)

print("\nTotal Sales Per Month:")
print(total_sales_per_month)

# Step 3: Calculate Average Growth Rate per Category
average_growth_rate = sales_df.pct_change().mean(axis=0) * 100

print("\nAverage Growth Rate Per Category (%):")
print(average_growth_rate)

# Step 4: Add Columns for Total Monthly Sales and Growth Rate
sales_df['Total Sales'] = total_sales_per_month
sales_df['Growth Rate'] = sales_df['Total Sales'].pct_change() * 100

print("\nUpdated DataFrame with Total Sales and Growth Rate:")
print(sales_df.head())

# Step 5: Apply Randomized Discount Based on Roll Number
discounted_df = sales_df.copy()
if random_seed % 2 == 0:
    discounted_df['Electronics'] *= 0.9  # Apply a 10% discount to Electronics
else:
    discounted_df['Clothing'] *= 0.85   # Apply a 15% discount to Clothing

print("\nDiscounted DataFrame (First 5 Rows):")
print(discounted_df.head())

# Part III: Visualizations (Matplotlib and Seaborn)

# Step 1: Plot Monthly Sales Trends by Category
plt.figure(figsize=(10, 6))
for category in categories:
    plt.plot(months, sales_df[category], label=category)
plt.title('Monthly Sales Trends by Category')
plt.xlabel('Month')
plt.ylabel('Sales')
plt.legend()
plt.grid(True)
plt.show()

# Step 2: Box Plot for Sales Distribution by Category
plt.figure(figsize=(10, 6))
sns.boxplot(data=sales_df[categories])
plt.title('Sales Distribution by Category')
plt.xlabel('Category')
plt.ylabel('Sales')
plt.show()

```

```python
# Q2: Array Manipulations

array = np.array([[1, -2, 3], [-4, 5, -6]])

# i. Element-wise Absolute Value
absolute_values = np.abs(array)
print("Element-wise Absolute Values:")
print(absolute_values)

# ii. Percentiles of Flattened Array
flattened = array.flatten()
percentiles = {
    '25th': np.percentile(flattened, 25),
    '50th': np.percentile(flattened, 50),
    '75th': np.percentile(flattened, 75)
}
print("\nPercentiles of Flattened Array:")
print(percentiles)

# iii. Statistical Measures (Mean, Median, Standard Deviation)
mean_flattened = np.mean(flattened)
median_flattened = np.median(flattened)
std_flattened = np.std(flattened)

mean_columns = np.mean(array, axis=0)
median_columns = np.median(array, axis=0)
std_columns = np.std(array, axis=0)

mean_rows = np.mean(array, axis=1)
median_rows = np.median(array, axis=1)
std_rows = np.std(array, axis=1)

print("\nStatistical Measures for Flattened Array:")
print("Mean:", mean_flattened)
print("Median:", median_flattened)
print("Standard Deviation:", std_flattened)

print("\nStatistical Measures per Column:")
print("Mean:", mean_columns)
print("Median:", median_columns)
print("Standard Deviation:", std_columns)

print("\nStatistical Measures per Row:")
print("Mean:", mean_rows)
print("Median:", median_rows)
print("Standard Deviation:", std_rows)

```

```python
# Q3: Floor, Ceiling, and Truncated Values

a = np.array([-1.8, -1.6, -0.5, 0.5, 1.6, 1.8, 3.0])

floor_values = np.floor(a)
ceiling_values = np.ceil(a)
truncated_values = np.trunc(a)

print("\nFloor Values:")
print(floor_values)

print("\nCeiling Values:")
print(ceiling_values)

print("\nTruncated Values:")
print(truncated_values)
```

```python
# Q4: Swap Two Elements in a List Using Temporary Variable

def swap_with_temp(lst, i, j):
    temp = lst[i]
    lst[i] = lst[j]
    lst[j] = temp
    return lst

lst = [1, 2, 3, 4, 5]
swapped_list = swap_with_temp(lst.copy(), 1, 3)
print("\nSwapped List:")
print(swapped_list)

```

```python
# Q5: Swap Two Elements in a Set

def swap_in_set(s, i, j):
    lst = list(s) # Convert set to list for swapping elements
    lst[i], lst[j] = lst[j], lst[i] # Swap elements at indices i and j
    return set(lst) # Convert back to set

s = {1, 2, 3, 4}
swapped_set = swap_in_set(s.copy(), list(s).index(2), list(s).index(4))
print("\nSwapped Set:")
print(swapped_set)

```
