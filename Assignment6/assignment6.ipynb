```python

import numpy as np
import matplotlib.pyplot as plt

M = float(input("Enter a value for M: "))
print("You entered the value: ", M)
x = np.linspace(-10, 10, 100)
y1 = M * (x**2)
y2 = M * np.sin(x)

plt.figure(figsize=(8, 5))
plt.plot(x, y1, label="M*x^2", color='b', linestyle='--')
plt.plot(x, y2, label="M*sin(x)", color='r', linestyle='-')

plt.title("Function Plots")
plt.xlabel("X-axis")
plt.ylabel("Y-axis")
plt.legend()
plt.grid(True)
plt.show()
```

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

data = {
    'Subject': ['Math', 'Science', 'English', 'History', 'Geography'],
    'Score': [85, 90, 78, 88, 92]
}

df = pd.DataFrame(data)

plt.figure(figsize=(10, 6))
bar_plot = sns.barplot(x='Subject', y='Score', hue='Subject', data=df, palette='Set2')

for index, row in df.iterrows():
    bar_plot.text(row.name, row.Score + 1, round(row.Score, 2), color='black', ha="center")

plt.title('Scores by Subject')
plt.xlabel('Subject')
plt.ylabel('Score')
plt.grid(True)
plt.show()
```

```python


import numpy as np
import matplotlib.pyplot as plt

roll_number = int(input('Enter your roll number: '))
np.random.seed(roll_number)

data = np.random.randn(50)

cumsum_data = np.cumsum(data)

fig, axes = plt.subplots(2, 2, figsize=(10, 8))

# Lineplot
axes[0, 0].plot(cumsum_data, color='b')
axes[0, 0].set_title('Cumulative Sum')
axes[0, 0].grid(True)

# Scatterplot 
axes[0, 1].scatter(range(50), data, color='r')
axes[0, 1].set_title('Scatter Plot (Random Noise)')
axes[0, 1].grid(True)

# Histogram 
axes[1, 0].hist(data, bins=10, color='g', edgecolor='black')
axes[1, 0].set_title('Histogram of Random Values')
axes[1, 0].grid(True)

# Boxplot
axes[1, 1].boxplot(data)
axes[1, 1].set_title('Boxplot of Random Values')
axes[1, 1].grid(True)
plt.tight_layout()
plt.show()

```

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

df = pd.read_csv('https://raw.githubusercontent.com/AnjulaMehto/MCA/main/company_sales_data.csv')

plt.figure(figsize=(10, 6))
sns.lineplot(x='month_number', y='total_profit', data=df, marker='o', color='b')
plt.title('Total Profit per Month')
plt.xlabel('Month Number')
plt.ylabel('Total Profit')
plt.grid(True)
plt.show()

plt.figure(figsize=(12, 8))
sns.lineplot(x='month_number', y='facecream', data=df, marker='o', label='Face Cream ')
sns.lineplot(x='month_number', y='facewash', data=df, marker='o', label='Face Wash')
sns.lineplot(x='month_number', y='toothpaste', data=df, marker='o', label='Toothpaste')
sns.lineplot(x='month_number', y='bathingsoap', data=df, marker='o', label='Bathing Soap')
sns.lineplot(x='month_number', y='shampoo', data=df, marker='o', label='Shampoo')
sns.lineplot(x='month_number', y='moisturizer', data=df, marker='o', label='Moisturizer')
plt.title('Product Sales Data')
plt.xlabel('Month Number')
plt.ylabel('Sales Units')
plt.legend()
plt.grid(True)
plt.show()

plt.figure(figsize=(10, 6))
sns.barplot(x='month_number', y='total_units', data=df, palette='Set3')
plt.title('Total Units Sold per Month')
plt.xlabel('Month Number')
plt.ylabel('Total Units Sold')
plt.grid(True)
plt.show()
```
