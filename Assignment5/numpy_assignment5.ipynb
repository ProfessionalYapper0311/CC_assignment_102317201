```python

import numpy as np
import matplotlib.pyplot as plt
```

```python

gfg = np.matrix([[4, 1, 9], [2, 3, 1], [4, 5, 6]])

print("Array is: ", gfg)
print("")
print("Sum of all elements:", np.sum(gfg))
print("")
print("Sum of all elements row-wise:", np.sum(gfg, axis=1))
print("")
print("Sum of all elements column-wise:", np.sum(gfg, axis=0))
```

```python

array = np.array([10, 52, 62, 16, 16, 54, 453])

print("Original array:", array)
print("")
print("Sorted array:", np.sort(array))
print("")
print("Indices of sorted array:", np.argsort(array))
print("")
print("Smallest 4 elements:", np.sort(array)[:4])
print("")
print("Largest 5 elements:", np.sort(array)[-5:])
```

```python

array = np.array([1.0, 1.2, 2.2, 2.0, 3.0, 2.0])

print("Array is:", array)
print("")
int_elements = array[array == array.astype(int)]
print("Integer elements:", int_elements)
print("")
float_elements = array[array != array.astype(int)]
print("Float elements:", float_elements)
```

```python

X = ord('R') + ord('B')

sales = np.array([X, X+50, X+100, X+150, X+200])
print("Sales data: ", sales)
```

```python

tax_rate = ((X % 5) + 5) / 100
taxes = sales * tax_rate

print("Tax rate is:", tax_rate)
print("")
print("Taxes applied on sales: ", taxes)
```

```python

discount = np.where(sales < X+100, sales * 0.05, sales * 0.10)

print("Discount applied based on sales: ", discount)
```

```python

sales_matrix = np.tile(sales, (3, 1))
sales_matrix = sales_matrix * np.array([1, 1.02, 1.02**2]).reshape(3, 1)

print("Sales for multiple weeks: ", sales_matrix)
```

```python
x = np.linspace(-10, 10, 100)

y1 = x**2
y2 = np.sin(x)
y3 = np.exp(x)
y4 = np.log(np.abs(x) + 1)

plt.figure(figsize=(10, 6))

plt.subplot(2, 2, 1)
plt.plot(x, y1, label="Y = x^2", color='blue')
plt.title("Parabola:- (Y = x^2)")
plt.xlabel("x")
plt.ylabel("y")
plt.legend()
plt.grid(True)

plt.subplot(2, 2, 2)
plt.plot(x, y2, label="Y = sin(x)", color='green')
plt.title("Sin:- (Y = sin(x))")
plt.xlabel("x")
plt.ylabel("y")
plt.legend()
plt.grid(True)

plt.subplot(2, 2, 3)
plt.plot(x, y3, label="Y = e^x", color='red')
plt.title("Exponential:- (Y = e^x)")
plt.xlabel("x")
plt.ylabel("y")
plt.legend()
plt.grid(True)

plt.subplot(2, 2, 4)
plt.plot(x, y4, label="Y = log(|x| + 1)", color='violet')
plt.title("Log:- (Y = log(|x| + 1))")
plt.xlabel("x")
plt.ylabel("y")
plt.legend()
plt.grid(True)

plt.suptitle("Functions Graph plot")

plt.tight_layout(rect=[0, 0, 1, 0.96])
plt.show()
```
